# Data Cleansing

```python
def clean_price(dat):
    dat = dat.apply(lambda x: x.replace("$",""))
    dat= dat.apply(lambda x: float(x.replace(",","")))
    return dat
```

## Buzzwords for Name

```python
def find_buzzword(col_name,num, asc=True):
    buzzes = []
    count_vect = CountVectorizer()
    temp = pd.DataFrame(full_dat[col_name],full_dat['price']).reset_index()
    temp = temp.dropna()
    X_train_counts = count_vect.fit_transform(temp[col_name])
    rf = RandomForestRegressor()
    rf.fit(X_train_counts,temp['price'])
    feature_list = pd.DataFrame(rf.feature_importances_).sort(columns=[0],ascending=False).index[:num]
    for word, index in count_vect.vocabulary_.iteritems():
        if index in feature_list:
            print word
            buzzes.append(word)
    return buzzes
```

```python
name_buzz = find_buzzword('name',50)
```

```python
transit_buzz = find_buzzword('transit',50)
```

```python
house_buzz = find_buzzword('house_rules',50,True)
```

## Create Indicator Variables for Missingness

```python
full_dat['name_missing'] = full_dat['name'].isnull()
full_dat['summary_missing'] = full_dat['summary'].isnull()
full_dat['space_missing'] = full_dat['space'].isnull()
full_dat['description_missing'] = full_dat['description'].isnull()
full_dat['notes_missing'] = full_dat['name'].isnull()
full_dat['transit_missing'] = full_dat['name'].isnull()
full_dat['house_rules_missing'] = full_dat['house_rules'].isnull()
full_dat['host_about_missing'] = full_dat['host_about'].isnull()
full_dat['square_feet_missing'] = full_dat['square_feet'].isnull()
full_columns += ['name_missing','summary_missing','space_missing','description_missing','notes_missing','transit_missing',
                   'house_rules_missing','host_about_missing','square_feet_missing']
```

```python
full_columns
```

```python
full_dat = full_dat[full_columns].fillna( 0)
```

# Transform Qualitative to Quantitative

```python
len(full_dat)
```

```python
house_quant = np.zeros(len(full_dat))
transit_quant = np.zeros(len(full_dat))
name_quant = np.zeros(len(full_dat))

for i in range(len(full_dat)):
    for j in range(50):
        try:
            if full_dat['house_rules'][i] != 0 and house_buzz[j] in full_dat['house_rules'][i] :
                house_quant[i] += j*full_dat['house_rules'][i].count(house_buzz[j])
            if full_dat['transit'][i] != 0 and transit_buzz[j] in full_dat['transit'][i]:
                transit_quant[i] += j*full_dat['transit'][i].count(transit_buzz[j])
            if full_dat['name'][i] != 0 and name_buzz[j] in full_dat['name'][i]:
                name_quant[i] += j*full_dat['name'][i].count(name_buzz[j])
        except: 
            pass
```

```python
house_quant=[-1 if x==0 else x for x in house_quant]
transit_quant=[-1 if x==0 else x for x in transit_quant]
name_quant=[-1 if x==0 else x for x in name_quant]
```

```python
full_dat['house_quant']= house_quant
full_dat['transit_quant'] = transit_quant
full_dat['name_quant'] = name_quant
```

```python
temp['price'] = clean_price(temp['price'])
temp['log_price'] = temp['price'].apply(lambda x: np.log(x))
```

```python
categorical = []
for i in range(len(temp['price'])):
    if temp['price'][i] < 100:
        categorical.append(1)
    elif temp['price'][i] < 200:
        categorical.append(2)
    else:
        categorical.append(3)
temp['categorical'] = categorical
```

# Fit the Models

## Random Forest Classifier

```python
X_train, X_test, y_train, y_test = train_test_split(train_test, temp['categorical'], test_size=0.33, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators = 500)
```

```python
rf.fit(X_train,y_train)
rf.score(X_test,y_test)
```

```python
svm = svm.SVC(C = .1)
```

```python
svm.fit(X_train,y_train)
svm.score(X_test,y_test)
```

```python
lr = LogisticRegression(C=.1)
```

```python
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```
